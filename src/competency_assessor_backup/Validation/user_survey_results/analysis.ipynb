```python
import pandas as pd

# Load your Excel file
df = pd.read_excel("Copy of Systems Engineering Competency Assessment Questionnaire Validation (Responses).xlsx")

# Group columns by category using a filter on column names
ambiguity_cols = [col for col in df.columns if col.startswith("Ambiguity:")]
complexity_cols = [col for col in df.columns if col.startswith("Complexity:")]
doublebarrel_cols = [col for col in df.columns if "Double‑barrelled" in col]
```

# Find ambiguity, complexity and double barreled percentages

```python
def compute_percentage(cols):
    total_yes = 0
    total = 0
    for col in cols:
        # Modify this logic if your affirmative response is different (e.g., "True", 1, etc.)
        total_yes += (df[col] == "Yes").sum()
        total += df[col].notnull().sum()  # count non-missing responses
    return (total_yes / total) * 100 if total > 0 else 0

ambiguity_percentage = compute_percentage(ambiguity_cols)
complexity_percentage = compute_percentage(complexity_cols)
doublebarrel_percentage = compute_percentage(doublebarrel_cols)

print("Ambiguity Percentage:", ambiguity_percentage)
print("Complexity Percentage:", complexity_percentage)
print("Double‑barrelled Percentage:", doublebarrel_percentage)
```

```python
import matplotlib.pyplot as plt

categories = ['Ambiguity', 'Complexity', 'Double‑barrelled']
percentages = [ambiguity_percentage, complexity_percentage, doublebarrel_percentage]

plt.bar(categories, percentages)
plt.ylabel("Percentage of 'Yes' responses")
plt.title("Survey Analysis")
plt.show()
```

# Identifying problematic questions

```python
import pandas as pd

# Load your Excel file. Replace 'your_survey_responses.xlsx' with your actual file name.
df = pd.read_excel("Copy of Systems Engineering Competency Assessment Questionnaire Validation (Responses).xlsx")

# Group columns based on keywords in the column names.
ambiguity_cols = [col for col in df.columns if col.startswith("Ambiguity:")]
complexity_cols = [col for col in df.columns if col.startswith("Complexity:")]
doublebarrel_cols = [col for col in df.columns if "Double‑barrelled" in col]

# Define a function to compute the percentage of "Yes" responses for each column.
def compute_problematic_questions(cols, threshold=0.5):
    problematic = {}
    for col in cols:
        # Modify this if your affirmative responses are coded differently (e.g., True, 1, etc.)
        yes_count = (df[col] == "Yes").sum()
        total_responses = df[col].notna().sum()  # Count non-missing responses
        if total_responses > 0:
            percentage = yes_count / total_responses
            if percentage >= threshold:
                problematic[col] = percentage * 100  # Convert to percentage
    return problematic

# Compute problematic questions for each category using a threshold of 50% (0.5)
problematic_ambiguity = compute_problematic_questions(ambiguity_cols, threshold=0.5)
problematic_complexity = compute_problematic_questions(complexity_cols, threshold=0.5)
problematic_doublebarrel = compute_problematic_questions(doublebarrel_cols, threshold=0.5)

# Print the results
print("Questions with high ambiguity issues:")
for question, perc in problematic_ambiguity.items():
    print(f"{question}: {perc:.1f}% Yes responses")

print("\nQuestions with high complexity issues:")
for question, perc in problematic_complexity.items():
    print(f"{question}: {perc:.1f}% Yes responses")

print("\nQuestions with high double‑barrelled issues:")
for question, perc in problematic_doublebarrel.items():
    print(f"{question}: {perc:.1f}% Yes responses")
```
