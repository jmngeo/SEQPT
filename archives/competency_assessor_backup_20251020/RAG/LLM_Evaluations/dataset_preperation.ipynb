```python
import os
import pandas as pd
import time
from dotenv import load_dotenv
from typing import Literal
from pydantic import BaseModel, Field
from langchain_openai import AzureChatOpenAI
import getpass
from langchain.schema import SystemMessage, HumanMessage

# Step 1: Load environment variables from .env file
load_dotenv()
```

```python
if not os.environ.get("AZURE_OPENAI_API_KEY"):
  os.environ["AZURE_OPENAI_API_KEY"] = getpass.getpass("Enter API key for Azure: ")
  
# Step 2: Initialize AzureChatOpenAI client with env vars
llm = AzureChatOpenAI(
    azure_endpoint=os.environ["AZURE_OPENAI_ENDPOINT"],
    azure_deployment="gpt-4o-mini",
    openai_api_version="2024-05-01-preview",
)



class ClassificationResult(BaseModel):
    is_systems_engineering: Literal["Yes", "No"] = Field(description="Whether the job description is related to Systems Engineering.")
    llm_classified_role_cluster: str = Field(description="The role cluster that best describes this job description. Use one of the predefined categories or 'None' if it doesn't fit any cluster.")

structured_llm = llm.with_structured_output(ClassificationResult)
```

```python
# Add a new column for classification results if not already present
if "is_systems_engineering" not in df.columns:
    df["is_systems_engineering"] = None

if "llm_classified_role_cluster" not in df.columns:
    df["llm_classified_role_cluster"] = None
    
# Output file to save progress
output_file = "classified_jobs_partial.json"

# Helper function to save DataFrame as JSON
def save_to_json(df, output_path):
    df.to_json(output_path, orient="records", indent=2)
```

```python
display(df["description"].count())
```

```python

# Step 4: Updated processing loop with expanded prompts
for index, row in df.iterrows():
    if pd.notnull(row.get("is_systems_engineering")):  # Skip already processed rows
        continue

    # Define the system message
    system_message = SystemMessage(
        content="""
        You are an assistant that classifies job descriptions from LinkedIn. Your tasks are:
        1. Determine whether the job is related to Systems Engineering (Systems Engineering is a transdisciplinary and integrative approach to enable the successful realization, use, and retirement of engineered systems, using systems principles and concepts, and scientific, technological, and management methods.).
        2. Identify which role cluster (from a predefined list) the job best fits into.
        """
    )

    # Define the user message
    user_message = HumanMessage(
        content=f"""
        Analyze the following job description and identify if this is a Systems Engineer Job. 

        For your information, the below are some of the categories that we could group Systems Engineer job roles into: 
        
        1. Customer: Represents the party that orders or uses a service (e.g., development order). The customer has influence on the design/technical execution of the system.
        2. Customer Representative: Forms the interface between the customer and the company. The roles in this cluster form the voice for all customer-relevant information required for the project.
        3. Project Manager: Is responsible for the planning and coordination on the project side. The roles assume responsibility for achieving the project goals and monitoring the resources (time, costs, personnel) within a time-limited framework and also have a moderating role in conflicts and disputes.
        4. System Developer: Has the overview from requirements to the decomposition of the system to the interfaces and the associated system elements (external to the system environment and internal between the elements). The system developer is responsible for integration planning and consults with the appropriate subject matter experts.
        5. Specialist Developer: Includes the various specialist areas, e.g., software, hardware, etc. They develop new technologies or realize the product/system on the basis of specifications from the system developer cluster.
        6. Production Planner/Coordinator: Takes on the preparation of the product realization and the transfer to the customer.
        7. Production Employee: Comprises the processes that are to be assigned to the implementation, assembly, and manufacture of the product through to goods issue and shipping. The individual system components are integrated into the overall system and verified with regard to their functionality.
        8. Quality Engineer/Manager: Ensures that the company’s quality standards are maintained in order to keep customer satisfaction high and ensure long-term competitiveness in the market. Close cooperation with the V&V operator, e.g., for the analysis of customer complaints and identification of the cause.
        9. Verification and Validation (V&V) Operator: Covers the topics of system verification & validation. The involvement of this role cluster in the early phases of system development can ensure that the system is verifiable and validatable.
        10. Service Technician: Deals with all service-related tasks at the customer’s site, i.e., installation, commissioning, professional training of users, as well as classic service tasks such as maintenance and repairs, or the area of after-sales.
        11. Process and Policy Manager: Is divided into a strategic and an operational level: On a strategic level, the process owner serves to develop internal guidelines in the development and creation or revision of process flows. On an operational level, the policy owner controls compliance with policies, laws and framework conditions that must be taken into account and fulfilled.
        12. Internal Support: Represents the advisory and supporting side during the development process within the project. This includes IT support, Qualification support, and Systems Engineering (SE) support.
        13. Innovation Management: Focuses on the commercially successful implementation of products or services, but also new business models or processes.
        14. Management: Forms the group of decision-makers and is represented by the management or department management. The cluster keeps an eye on the company’s goals, visions, and values. Since the opinion of the cluster is crucial for project progress, management is an important stakeholder in every respect.


        Job Description: {row['description']}
        """
    )

    # Combine messages
    messages = [system_message, user_message]

    try:
        # Invoke the LLM with structured output
        result = structured_llm.invoke(messages)  # This directly returns a ClassificationResult object

        # Access attributes directly from the result object
        classification = result.is_systems_engineering
        role_cluster = result.llm_classified_role_cluster

    except Exception as e:
        print(f"Unexpected error: {e}")
        classification = "Error"
        role_cluster = "Error"

    # Update DataFrame with results
    df.at[index, "is_systems_engineering"] = classification
    df.at[index, "llm_classified_role_cluster"] = role_cluster


    # Save progress
    save_to_json(df, output_file)

    # Wait to handle rate limits
    time.sleep(12)

# Save final classified subset
save_to_json(df, "classified_jobs_subset_final.json")
print("Classification for subset completed!")
```

# Classify the job description into: Responsible for, Supporting and Designing

```python
import pandas as pd
from typing import List
from pydantic import BaseModel, Field
from langchain_openai import AzureChatOpenAI

# Step 1: Define the Pydantic model for structured output
class JobTasks(BaseModel):
    responsible_for: List[str] = Field(description="Tasks the user is responsible for.")
    supporting: List[str] = Field(description="Tasks the user is supporting.")
    designing: List[str] = Field(description="Tasks the user is designing.")

# Step 2: Initialize the LLM
llm = AzureChatOpenAI(
    azure_endpoint=os.environ["AZURE_OPENAI_ENDPOINT"],
    azure_deployment="gpt-4o-mini",
    openai_api_version="2024-05-01-preview",
    temperature = 0
)

structured_llm = llm.with_structured_output(JobTasks)

# Step 3: Iterate through the DataFrame and get structured output
results = []
```

```python
# Step 3: Load the filtered DataFrame
classified_file_name = "job_description_systems_engineer_sampled.json"
df_filtered = pd.read_json(classified_file_name)
```

```python
display(df_filtered)
```

```python

# Initialize progress file
progress_file = "llm_processed_jobs_to_task.json"

# Check if progress file exists to resume
if os.path.exists(progress_file):
    with open(progress_file, "r") as f:
        processed_results = json.load(f)
else:
    processed_results = []

# Process already completed indices
processed_indices = {res["index"] for res in processed_results}

# Step 4: Iterate through the DataFrame and get structured output
for index, row in df_filtered.iterrows():
    if index in processed_indices:
        continue  # Skip already processed rows

    description = row['description']

    # Create the prompt
    prompt = f"""
    This is a job description from LinkedIn: "{description}"

    Read the job description and extract the following information:
    - Tasks the candidate is responsible for.
    - Tasks the candidate is supporting.
    - Tasks the candidate is designing.
    """

    try:
        # Invoke the LLM and get the structured response
        response = structured_llm.invoke(prompt)

        # Append the structured response to the processed results
        processed_results.append({
            "index": index,
            "responsible_for": response.responsible_for,
            "supporting": response.supporting,
            "designing": response.designing
        })

    except Exception as e:
        print(f"Error processing row {index}: {e}")
        processed_results.append({
            "index": index,
            "responsible_for": [],
            "supporting": [],
            "designing": []
        })

    # Save progress after each row
    with open(progress_file, "w") as f:
        json.dump(processed_results, f, indent=2)
    
    time.sleep(15)

# Step 5: Add the results back to the DataFrame
# Convert processed results to a DataFrame and merge with df_filtered
processed_df = pd.DataFrame(processed_results)
df_filtered = df_filtered.merge(processed_df, left_index=True, right_on="index", how="left")

# Save the final updated DataFrame
df_filtered.to_json("updated_jobs_with_tasks.json", orient="records", indent=2)

# Display the final DataFrame
print(df_filtered.head())
```

```python
# Step 1: Connect to PostgreSQL and fetch ISO processes
def fetch_iso_processes():
    connection = psycopg2.connect(
        dbname=os.environ["DB_DB"],
        user=os.environ["DB_USER"],
        password=os.environ["DB_PASSWORD"],
        host=os.environ["DB_HOST"],
        port="5432"
    )
    query = "SELECT name, description FROM iso_processes"
    iso_processes_df = pd.read_sql(query, connection)
    connection.close()
    return iso_processes_df.to_dict(orient="records")

iso_processes = fetch_iso_processes()
```
